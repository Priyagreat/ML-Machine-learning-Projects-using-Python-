```python
import pandas as pd
df=pd.read_csv('drive/MyDrive/Colab Notebooks/carprices.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Car Model'],df['Sell Price($)'])
```

```python
dummy = pd.get_dummies(df['Car Model'])
dummy
```

```python
merged = pd.concat([df,dummy],axis='columns')
merged
```

```python
final =merged.drop(['Car Model','BMW X5'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('Sell Price($)', axis='columns')
x
y = final['Sell Price($)']
y
model.fit(x, y)
```

```python
model.fit(x, y)
```

```python
##Price of mercedez benz that is 4 yr old with mileage 45000

model.predict([[45000,4,0,1]])
```

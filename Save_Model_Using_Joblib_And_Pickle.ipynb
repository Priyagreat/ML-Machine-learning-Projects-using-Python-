```python
#y=mx+b , price(dependent variable)=m*area( independent variable)+b
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/homeprices.csv')
print(df.head(5))
plt.scatter(df.area,df.price,marker='+',color='red')
plt.xlabel('area')
plt.ylabel('price',color='red')
plt.show()




```

```python
rg=LinearRegression()
rg.fit(df[['area']],df.price)
plt.scatter(df.area,df.price,marker='+',color='red')
plt.xlabel('area',fontsize=20)
plt.ylabel('price',color='red', fontsize=20)
plt.plot(df.area,rg.predict(df[['area']]),color='blue')
plt.show()
```

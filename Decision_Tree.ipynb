```python
import pandas as pd
df=pd.read_csv('drive/MyDrive/Colab Notebooks/salaries.csv')
df
```

```python
input= df.drop('salary_more_then_100k', axis='columns')
target = df['salary_more_then_100k']
input
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
input['company_n'] = le_company.fit_transform(input['company'])
input['job_n'] = le_job.fit_transform(input['job'])
input['degree_n'] = le_degree.fit_transform(input['degree'])
input.head()
```

```python
inputs_n = input.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
model.score(inputs_n,target)
```

```python
##Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?

model.predict([[2,1,0]])
```

```python
##Is salary of Google, Computer Engineer, Masters degree > 100 k ?
model.predict([[2,1,1]])
```

```python
import numpy as np
import pandas as pd

def gradient_descent(X, y):
    m_current = n_current = 0
    iteration = 100
    n = len(X)
    learning_rate = 0.01

    for i in range(iteration):
        y_prediction = m_current * X + n_current
        cost= (1/n)*sum([val**2 for val in (y-y_prediction)])
        md = -(2/n)*sum(X*(y-y_prediction))
        nd = -(2/n)*sum(y-y_prediction)
        m_current = m_current - learning_rate * md
        n_current = n_current - learning_rate * nd
        print("m {}, n {},cost {}".format(m_current, n_current, cost,i))

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```
